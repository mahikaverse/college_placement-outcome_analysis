```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("placementdata.csv")
data.head()
data.info()
data.describe()
data.isnull().sum()

```

```python
data["PlacementStatus"] = data["PlacementStatus"].map({
    "Placed": 1,
    "NotPlaced": 0
})

data["PlacementTraining"] = data["PlacementTraining"].map({
    "Yes": 1,
    "No": 0
})

data["ExtracurricularActivities"] = data["ExtracurricularActivities"].map({
    "Yes": 1,
    "No": 0
})


```

```python
data["PlacementStatus"].value_counts()
data.groupby("PlacementStatus")["CGPA"].mean()
data.groupby("PlacementStatus")["Internships"].mean()
data.groupby("PlacementStatus")["PlacementTraining"].mean()

```

Visualization (Matplotlib)

```python
data["PlacementStatus"].value_counts().plot(kind="bar")
plt.title("Placement Distribution")
plt.show()
```

CGPA vs Placement

```python
data.groupby(pd.cut(data["CGPA"], bins=5))["PlacementStatus"].mean().plot(kind="bar")
plt.title("CGPA Range vs Placement Rate")
plt.ylabel("Placement Rate")
plt.show()

```

Internships vs Placement

```python
data.groupby("Internships")["PlacementStatus"].mean().plot(kind="bar")
plt.title("Internships vs Placement Rate")
plt.ylabel("Placement Rate")
plt.show()

```

Training vs Placement

```python
data.groupby("PlacementTraining")["PlacementStatus"].mean().plot(kind="bar")
plt.title("Training vs Placement Rate")
plt.show()
```

Soft Skills vs Placement

```python
data.groupby("SoftSkillsRating")["PlacementStatus"].mean().plot(kind="bar")
plt.title("Soft Skills Rating vs Placement Rate")
plt.ylabel("Placement Rate")
plt.show()

```

Aptitude vs Placement

```python
data.groupby(pd.cut(data["AptitudeTestScore"], bins=5))["PlacementStatus"].mean().plot(kind="bar")
plt.title("Aptitude Score Range vs Placement Rate")
plt.ylabel("Placement Rate")
plt.show()

```

Projects vs Placement

```python
data.groupby("Projects")["PlacementStatus"].mean().plot(kind="bar")
plt.title("Projects vs Placement Rate")
plt.ylabel("Placement Rate")
plt.show()

```
